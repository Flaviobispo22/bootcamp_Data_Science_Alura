```python
import pandas as pd
```

```python
dh = pd.read_json('/content/dados_hospedagem.json')
```

```python
dh.head()
```

```python
dh = pd.json_normalize(dh['info_moveis'])
dh
```

```python
colunas = list(dh.columns)
colunas
```

```python
dh = dh.explode(colunas[3:])
dh
```

```python
dh.reset_index(inplace=True, drop=True)
dh.head()
```

```python
dh.info()
```

```python
import numpy as np
```

```python
dh['max_hospedes'] = dh['max_hospedes'].astype(np.int64)
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dh[col_numericas] = dh[col_numericas].astype(np.int64)
```

```python
dh.info()
```

```python
dh['avaliacao_geral'] = dh['avaliacao_geral'].astype(np.float64)
```

```python
dh.info()
```

```python
dh['preco'] = dh['preco'].apply(lambda x : x.replace('$', '').replace(',','').strip())
```

```python
dh['preco'] = dh['preco'].astype(np.float64)
```

```python
dh.info()
```

```python
dh[['taxa_deposito','taxa_limpeza']] = dh[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
dh[['taxa_deposito','taxa_limpeza']] = dh[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dh.info()
```
